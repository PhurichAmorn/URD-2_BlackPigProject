```python
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO

# Load Model
model = YOLO('best_saved_model/best_float16.tflite', task='detect')

# Run inference
results = model.predict(source='vid1.png', imgsz=640, conf=0.4)


res_plotted = results[0].plot()

res_rgb = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)

# Display the result
plt.figure(figsize=(10, 10))
plt.imshow(res_rgb)
plt.axis('off')
plt.title("Pig Detection Test - TFLite Model")
plt.show()
```
